# Set plot def
# img[행, 열, 채널(1~9)]
# Ch1, 2 - 10.65 GHz
# Ch3, 4 - 18.70 GHz
# Ch5 - 23.80 GHz
# Ch6, 7 - 36.5 GHz
# Ch8, 9 - 89.0 GHz

```python
def weather9(img):
    fig, ax = plt.subplots(3,3,figsize=(15, 10))
    ax = ax.ravel()
    
    ax[0].set_title('Ch.1, 10.65 GHz: Liuid precip')
    ax[0].imshow(img[:,:,0])
    
    ax[1].set_title('Ch.2, 10.65 GHz: Liuid precip')
    ax[1].imshow(img[:,:,1])
    
    ax[2].set_title('Ch.3, 18.70 GHz: Moderate to light precip')
    ax[2].imshow(img[:,:,2])
    
    ax[3].set_title('Ch.4, 18.70 GHz: Moderate to light precip')
    ax[3].imshow(img[:,:,3])
    
    ax[4].set_title('Ch.5, 23.80 GHz: Water vapor emission')
    ax[4].imshow(img[:,:,4])
    
    ax[5].set_title('Ch.6, 36.5 GHz: Moderate to light precip')
    ax[5].imshow(img[:,:,5])
    
    ax[6].set_title('Ch.7, 36.5 GHz: Moderate to light precip')
    ax[6].imshow(img[:,:,6])
    
    ax[7].set_title('Ch.8, 89.0 GHz: Detection of ice particle')
    ax[7].imshow(img[:,:,7])
    
    ax[8].set_title('Ch.9, 89.0 GHz: Detection of ice particle')
    ax[8].imshow(img[:,:,8])
    
    plt.tight_layout()
```

# 채널별 주파수 및 관찰 목표 차이점 확인

```python
print('subset_010464_08.npy')
weather9(img)
```
